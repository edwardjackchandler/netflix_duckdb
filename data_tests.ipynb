```python
import duckdb
conn = duckdb.connect('new_db.duckdb')
# print(conn.execute('''
# select * from netflix_titles limit 1;
# ''').fetchall())
print(conn.execute('''
show tables;
''').fetchall())
```

```python
conn.execute("""
SELECT
    strptime('August 5, 2019', '%B %d, %Y') AS prem_date,
    *
FROM
    src_netflix_originals_imdb
""").df()
```

```python
conn.execute("""
SELECT
    CASE
        WHEN show_type = 'Movie' THEN CAST(LEFT(duration_string, -3) AS INT)
        ELSE NULL
    END AS movie_length,
    CASE
        WHEN show_type = 'TV Show' THEN CAST(LEFT(duration_string, -7) AS INT)
        ELSE NULL
    END AS tv_show_seasons, 
FROM
    src_netflix_titles
""").df()
```
